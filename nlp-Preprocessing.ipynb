```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

# Téléchargez la liste des mots vides
nltk.download('stopwords')
nltk.download('punkt')

# Charger les données dans un DataFrame
data = pd.read_csv('/kaggle/input/customer-support-on-twitter/sample.csv', usecols=['tweet_id', 'text','text'])

# Mise en minuscules
data['text_preprocessing'] = data['text'].str.lower()
# Afficher les premières lignes du DataFrame après nettoyage
print(data.head())
```

```python
# Suppression des ponctuations
data['text_preprocessing'] = data['text_preprocessing'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
print(data.head())
```

```python
#supprime les mots tel que "is" ,"the" , "le" , "un", ... etc
stop_words = set(stopwords.words('english'))
data['text_preprocessing'] = data['text_preprocessing'].apply(lambda x: ' '.join([word for word in word_tokenize(x) if word.lower() not in stop_words]))
print(data.head())
```

```python
#suppression des mots frequent
from collections import Counter

# Calculer la fréquence de chaque mot dans le text
word_freq = Counter()
for text in data['text_preprocessing']:
    word_freq.update(text.split())

# Identifier les mots fréquents à supprimer 
SEUIL_FRÉQUENCE = 30
stopwords = set()
for word, freq in word_freq.items():
    if freq > SEUIL_FRÉQUENCE:  
        stopwords.add(word)

# Supprimer les mots fréquents de chaque texte
data['text_preprocessing'] = data['text_preprocessing'].apply(lambda x: ' '.join([word for word in x.split() if word not in stopwords]))
print(data.head())
```

```python
#suppression des mots rares : 
from collections import Counter

# Calculer la fréquence de chaque mot dans le text
word_freq = Counter()
for text in data['text_preprocessing']:
    word_freq.update(text.split())

# Identifier les mots fréquents à supprimer 
SEUIL_FRÉQUENCE_RARE = 1
mots_rares = set(word for word, freq in word_freq.items() if freq <= SEUIL_FRÉQUENCE_RARE)

# Supprimer les mots rares de chaque texte
data['text_preprocessing'] = data['text_preprocessing'].apply(lambda x: ' '.join([word for word in x.split() if word not in mots_rares]))
print(data.head())
```

```python
# Stemming : eating --> eat
ps = PorterStemmer()
data['text_preprocessing'] = data['text_preprocessing'].apply(lambda x: ' '.join([ps.stem(word) for word in word_tokenize(x)]))
print(data.head())
```

```python
# Suppression des émojis, des émoticônes, des URL , ....
data['text_preprocessing'] = data['text_preprocessing'].apply(lambda x: re.sub(r'\W', ' ', x))
print(data.head())
```

```python
data['text_cleaned'] = data['text_preprocessing']
```

```python
data_resultat = data[['tweet_id', 'text', 'text_cleaned']]
print(data_resultat)
```
